Risk Assessment

```python
# merging the two dataframes

# Full merge on the 'SK_ID_CURR'
Full_merged_df = pd.merge(Filtered_application_data, Prev_application, on='SK_ID_CURR', how='inner')

# Display the result
Full_merged_df
```

```python
Rejection_reasons = Full_merged_df.groupby('CODE_REJECT_REASON')['NAME_INCOME_TYPE'].value_counts(normalize=True).unstack()*100
Rejection_reasons
```

Assessment by Job Type

```python
# converting NaNs to Zeroes
Rejection_reasons = Rejection_reasons.fillna(0)
Rejection_reasons
```

```python
# Increasing the size of the chart
fig, ax = plt.subplots(figsize=(10, 6)) 

# Creating a clustered bar chart
Rejection_chart = Rejection_reasons.plot(kind='barh', stacked=True, ax=ax, width= 0.8)

# Add labels and title
plt.ylabel('Rejection Reasons')
plt.xlabel('Count of Applicants')
plt.title('Rejection Reasons per Job Type')

# Add legend
Rejection_chart.legend(title='Income Type', bbox_to_anchor=(1, 1))

# Show the plot
plt.show()
```

Assessment by Credit Amount and Down Payment

```python
Down_payment_per_credit = Full_merged_df.groupby('NAME_CONTRACT_STATUS')[['AMT_CREDIT_x','AMT_DOWN_PAYMENT']].mean().round(2)
Down_payment_per_credit
```

```python
# Increasing the size of the chart
fig, ax = plt.subplots(figsize=(10, 6)) 

# Creating a clustered bar chart
Payment_chart = Down_payment_per_credit.plot(kind = 'barh', ax = ax, stacked= False)

# Add labels and title
plt.ylabel("Contract Status")
plt.xlabel('Average Amount')
plt.title('Previous Contract Approval per Average Credit')

# Add legend
Rejection_chart.legend(title='Average Amount', bbox_to_anchor=(1, 1))

# Show the plot
plt.show()

# Move the legend outside the box
Payment_chart.legend(loc='center left', bbox_to_anchor=(1, 0.5), title='Average Amount')
```
